#### Task 1: Create wireframe of hyperbolic paraboloid on the domain made from intervals `<-10,10>` x `<-10,10>`

* hyperbolic paraboloid equation is: z = x*x - y*y

```python
# calculate hyperbolic parabloid
x = np.linspace(-10,10,500)
y = np.linspace(-10,10,500)

# cartesian product
X,Y = np.meshgrid(x,y)

# distribution parameters
# x
mu_x = 0
variance_x = 10

# y
mu_y = 0
variance_y = 10

# a means list
means = [mu_x, mu_y]

# a covariance
cov = [[variance_x, 0], [0, variance_y]]

# a multivariate normal distribution
mn = multivariate_normal(means, cov)
```

```python
# wireframe
position = np.array([X,Y]).T

fig,ax = plt.subplots(figsize=(12.8,9.6))

ax = plt.axes(projection='3d')

Z = (X*X)-(Y*Y)

ax.plot_wireframe(X,Y,Z)
plt.show()
```
