# Pictures

```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
myImage = Image.open("data/flower_square.jpg")
plt.figure(figsize=(9, 6))
plt.imshow(myImage)
```

```python
imgmat = np.array(myImage)
plt.figure(figsize=(9,6))
plt.imshow(imgmat)
```

```python
print(type(myImage))
print(type(imgmat))
print(imgmat.shape)
```

```python
imgmat = np.array(myImage)
# storing the shape of the picture into the variables
lx, ly, lz = imgmat.shape

X,Y = np.ogrid[0:lx,0:ly]
print(X.shape)
print(Y.shape)
```

```python
mask = (X - lx / 2) ** 2 + (Y - ly / 2) ** 2 > lx * ly / 4

# Where mask is True, insert black color
imgmat[:,:,0][mask] = 0
imgmat[:,:,1][mask] = 0
imgmat[:,:,2][mask] = 0
```

```python
plt.figure(figsize=(9,6))
plt.imshow(imgmat)
```

```python
mask = (X - lx / 3) ** 2 + (Y - ly / 3) ** 2 > lx * ly / 4

# Where mask is True, insert black color
imgmat[:,:,0][mask] = 0
imgmat[:,:,1][mask] = 0
imgmat[:,:,2][mask] = 0

plt.figure(figsize=(9,6))
plt.imshow(imgmat)
```

```python
mask = (X - lx / 10) ** 2 + (Y - ly / 10) ** 2 > lx * ly / 4

# Where mask is True, insert black color
imgmat[:,:,0][mask] = 0
imgmat[:,:,1][mask] = 0
imgmat[:,:,2][mask] = 0

plt.figure(figsize=(9,6))
plt.imshow(imgmat)
```
